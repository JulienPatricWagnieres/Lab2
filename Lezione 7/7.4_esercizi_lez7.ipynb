### **1. Distribuzioni di probabilità**
Genera 1000 valori casuali da una distribuzione gamma con parametro di forma pari a 1.
Suggerimento: il parametro di forma viene passato come primo argomento quando si "congela" la distribuzione.

Traccia l’istogramma del campione e sovrapponi la PDF della distribuzione.

Stima il parametro di forma dal campione usando il metodo fit.

Extra:
Le distribuzioni hanno molti metodi utili. Esplorali usando il completamento automatico con il tasto TAB.

Traccia la funzione di distribuzione cumulativa (CDF).

Calcola la varianza.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
```

```python
dist = sp.stats.gamma(a=1)
sample = dist.rvs(size=1000)
plt.figure(figsize=(15,8))
plt.hist(sample,bins=50,density=True,label='campione normalizzato')
x = np.linspace(0,sample.max(),200)
plt.plot(x,dist.pdf(x),label='PDF')
plt.title(f'Gamma, size={sample.size}')
plt.legend()
plt.grid(True)
plt.xticks(range(0,8))
plt.show()

par = sp.stats.gamma.fit(sample)
print(par)

plt.figure(figsize=(15,8))
plt.title('Funzione cumulativa')
plt.plot(x,dist.cdf(x))
plt.axhline(y=1,color='red')
plt.show()

print('varianza teorica:',dist.var())
print('varianza campione:',sample.var())
print('variaza calcolata dal fit:',par[0]*par[2]**2)
```

### **2. Fitta i dati**
Prova a fittare i dati sottostanti con le migliori curve, calcola il MAE e l'RMSE
Esercizio 

```python
import numpy as np
temp_max = np.array([17, 19, 21, 28, 33, 38, 37, 37, 31, 23, 19, 18])
temp_min = np.array([-62, -59, -56, -46, -32, -18, -9, -13, -25, -46, -52, -58])
import matplotlib.pyplot as plt
import scipy as sp

```

```python
mesi= np.arange(1,13)
x_interpol = np.linspace(1,12,200)

fig, ax = plt.subplots(2,2,figsize=(18,18))
ax= ax.flatten()

def funz(x,a,b,c,d,e):
    return a*np.sin(b*x+c)+d*x+e

for i in range(1,4):
    coeffs = np.polyfit(mesi,temp_max,deg=i)
    y_interpol = np.polyval(coeffs,x_interpol)
    y_p = np.polyval(coeffs,mesi)
    mae = np.mean(abs(y_p - temp_max))
    rmse = np.sqrt(np.mean((y_p - temp_max)**2))

    ax[i-1].scatter(mesi,temp_max)
    ax[i-1].set_title(f'temp_max\nMAE:{mae}\nRMSE:{rmse}')
    ax[i-1].set_xlabel('mesi')
    ax[i-1].set_ylabel('Temp')


    ax[i-1].plot(x_interpol,y_interpol,label=f'grado:{i}')
    ax[i-1].legend()

param, _ = sp.optimize.curve_fit(funz,mesi,temp_max,p0=[10, 2*np.pi/12, 0, 0, np.mean(temp_max)])
y_pred = funz(x_interpol,*param)
ax[3].scatter(mesi,temp_max)
ax[3].plot(x_interpol,y_pred,label='fit seno')
ax[3].set_xlabel('mesi')
ax[3].set_ylabel('Temp')
y_p = funz(mesi,*param)
mae = np.mean(abs(y_p - temp_max))
rmse = np.sqrt(np.mean((y_p - temp_max)**2))
ax[3].set_title(f'temp_max\nMAE:{mae}\nRMSE:{rmse}')
ax[3].legend()


fig, bx = plt.subplots(2,2,figsize=(18,18))
bx= bx.flatten()
for i in range(1,4):
    coeffs = np.polyfit(mesi,temp_min,deg=i)
    y_interpol = np.polyval(coeffs,x_interpol)
    y_p = np.polyval(coeffs,mesi)
    mae = np.mean(abs(y_p - temp_min))
    rmse = np.sqrt(np.mean((y_p - temp_min)**2))

    bx[i-1].scatter(mesi,temp_min)
    bx[i-1].set_title(f'temp_min\nMAE:{mae}\nRMSE:{rmse}')
    bx[i-1].set_xlabel('mesi')
    bx[i-1].set_ylabel('Temp')

    bx[i-1].plot(x_interpol,y_interpol,label=f'grado={i}')
    bx[i-1].legend()


param, _ = sp.optimize.curve_fit(funz,mesi,temp_min,p0=[10, 2*np.pi/12, 0, 0, np.mean(temp_min)])
y_pred = funz(x_interpol,*param)
bx[3].scatter(mesi,temp_min)
bx[3].plot(x_interpol,y_pred,label='fit seno')
bx[3].set_xlabel('mesi')
bx[3].set_ylabel('Temp')
y_p = funz(mesi,*param)
mae = np.mean(abs(y_p - temp_min))
rmse = np.sqrt(np.mean((y_p - temp_min)**2))
bx[3].set_title(f'temp_min\nMAE:{mae}\nRMSE:{rmse}')
bx[3].legend()


plt.show()
```
